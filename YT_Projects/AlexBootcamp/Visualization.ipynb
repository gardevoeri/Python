<h1> Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight') # Changes the style of the charts
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
df.plot() # The default plot is 'line' but it can be called as an argument
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar') # Specifying the column creates a plot of those values only
```

```python
df.plot.barh(stacked = True) # Used for horizontal bar
```

```python
df['Flavor Rating'].plot.barh()
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 400, c = 'red' )
# s for Size and c for Color
```

```python
df.plot.hist()
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10,5))
# Specify a iterable size to increase the size of the picture
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,10))
```
