```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)

plt.rcParams['figure.figsize'] = (16,6)

plt.show()
```

```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
df2.plot()
```

```python
df3 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df3 = df3.transpose()
```

```python
df3.plot()
```
