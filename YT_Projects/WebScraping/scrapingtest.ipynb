```python
import requests
```

```python
html_text = requests.get("https://www.google.com/").text
html_text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_text)
soup
```

```python
divs = soup.findAll('div')
divs
```

```python
links = soup.findAll('a', class_ = 'gb1')
links
```
